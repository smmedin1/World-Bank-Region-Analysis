### Considering Mean or Median to replace the final null values

After imputing the null values and removing the 4 columns which have more than 75% of missing data. There are still some null values remaining in the dataset. 

We decided to use the normality testing where as general rule of thumb: If skewness is less than -1 or greater than 1, the distribution is highly skewed. If skewness is between -1 and -0.5 or between 0.5 and 1, the distribution is moderately skewed. If skewness is between -0.5 and 0.5, the distribution is approximately symmetric.

Therefore, if the skewness is more than 0.5 or less than 0.5 we choose median for those values and if the skewness is between -0.5 and 0.5 we choose mean.

```python
#creating a column named column_mean
column_mean =[]
#count_mean is to store the values if column have mean value
count_mean = 0

#creating new column for median values
column_median = []
#count median is to store values if columns have median values
count_median = 0

#for loop is used to check if values lie less than -0.5 to more than 0.5
for col in my_data.iloc[:,4:45]:
    if my_data[col].skew() >= 0.5 or my_data[col].skew() <= -0.5:
        column_median.append(col)
        count_median += 1
        
#else statement is used to append mean values        
    else:
        column_mean.append(col)
        count_mean += 1

#print the columns which have mean values 
print(f"**********\n\033[1m\033[94m Mean Columns\033[0;0m\033[0;0m\n**********")
count_mean

#this will print the column which lie in mean area.
for data in column_mean:
    print(data)

#print the columns which have median values
print(f"**********\n\033[1m\033[94m Median Columns\033[0;0m\033[0;0m\n**********")

#this will print the column which lie in median area.
count_median
for data in column_median: 
    print(data)
```
